```python
#BMI Calculator
name= input("What is your name? ")
height = float(input("What is your height in meters? "))
weight = float(input("What is your weight in kg? "))

bmi = weight / (height ** 2)

print(f"HI {name } , Your BMI is {bmi:.2f}")
```

<h1 align="center">📖 Dictionary</h1>

<p align="center">
Python Dictionary stores data as key-value pairs.
</p>

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
x = thisdict["model"]

print(x)
print(thisdict.get("year"))
```

```python
car = {
"brand": "Ford",
"model": "Mustang",
"year": 1964
}

x = car.keys()

print(x) #before the change

car["color"] = "white"

print(x) #after the change

print(car.keys())
print(car.values())
```
